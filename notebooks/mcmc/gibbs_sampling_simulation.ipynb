```python
import numpy as np
import matplotlib.pyplot as plt

#      
def sample_x_given_y(y):
    mean = -y / 2
    std_dev = 1 / np.sqrt(2)
    return np.random.normal(mean, std_dev)

def sample_y_given_x(x):
    mean = -x / 2
    std_dev = 1 / np.sqrt(2)
    return np.random.normal(mean, std_dev)

# Gibbs Sampling 
def gibbs_sampling(n_samples=5000):
    samples = np.zeros((n_samples, 2))
    x, y = 0, 0  #  

    for i in range(n_samples):
        # x 
        x = sample_x_given_y(y)
        # y 
        y = sample_y_given_x(x)
        #  
        samples[i] = [x, y]
    
    return samples

#  
np.random.seed(42)
samples = gibbs_sampling(n_samples=5000)

# 
plt.figure(figsize=(10, 5))
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.3, s=10, label="Gibbs Samples")
plt.title("Gibbs Sampling from Bivariate Distribution")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
