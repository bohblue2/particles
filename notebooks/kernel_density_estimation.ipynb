```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gaussian_kde

#   
np.random.seed(42)
data = np.random.normal(0, 1, size=500)  #   

# 1. KDE  
kde = gaussian_kde(data)
x = np.linspace(-4, 4, 1000)
kde_density = kde(x)

# 2.  MLE 
mu_mle = np.mean(data)  #  
sigma_mle = np.std(data, ddof=1)  #  
normal_density = norm.pdf(x, loc=mu_mle, scale=sigma_mle)

# 
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.5, label="Sample Histogram")
plt.plot(x, kde_density, label="KDE", linewidth=2)
plt.plot(x, normal_density, label=f"MLE Fit (Normal): ={mu_mle:.2f}, ={sigma_mle:.2f}", linestyle="--")
plt.legend()
plt.title("Density Estimation")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```
