```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gaussian_kde

# 샘플 데이터 생성
np.random.seed(42)
data = np.random.normal(0, 1, size=500)  # 정규분포에서 추출된 샘플

# 1. KDE로 밀도 추정
kde = gaussian_kde(data)
x = np.linspace(-4, 4, 1000)
kde_density = kde(x)

# 2. 정규분포에 MLE로 적합
mu_mle = np.mean(data)  # 평균 추정
sigma_mle = np.std(data, ddof=1)  # 표준편차 추정
normal_density = norm.pdf(x, loc=mu_mle, scale=sigma_mle)

# 시각화
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.5, label="Sample Histogram")
plt.plot(x, kde_density, label="KDE", linewidth=2)
plt.plot(x, normal_density, label=f"MLE Fit (Normal): μ={mu_mle:.2f}, σ={sigma_mle:.2f}", linestyle="--")
plt.legend()
plt.title("Density Estimation")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```
